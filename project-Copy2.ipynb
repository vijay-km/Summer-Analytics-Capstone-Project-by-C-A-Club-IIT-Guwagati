```python
## Include "object" data-types for categorical columns.
df = df.drop('date',axis='columns')
print(df)
df_cat = df.select_dtypes(include="object").columns.values

## Use One Hot encoding on the df and use columns as df_cat
df_encoded=pd.get_dummies(df,columns=df_cat)
```

```python
df_encoded.head()
```

```python
df2 = df_encoded.copy()
df2
```

```python
X = df2.drop(['revenue'],axis=1)

y = df2['revenue']
y
```

```python
# df1 = df2.drop('revenue', axis = 'columns')
# df1
```

```python
dt = dt.drop('date',axis='columns')
print(dt)
dt_cat = dt.select_dtypes(include="object").columns.values

## Use One Hot encoding on the df and use columns as df_cat
dt_encoded=pd.get_dummies(dt,columns=dt_cat)
dt2 = dt_encoded.copy()
dt2
```

```python
feature_difference = set(X) - set(dt2)

# create zero-filled matrix where the rows are equal to the number
# of row in `test` and columns equal the number of categories missing (i.e. set difference 
# between relevant `train` and `test` columns
feature_difference_df = pd.DataFrame(data=np.zeros((dt2.shape[0], len(feature_difference))),
                                     columns=list(feature_difference))

# add "missing" features back to `test
dt2 = dt2.join(feature_difference_df)
dt2
```

```python
dt2.columns
```

```python
feature_difference_2 = set(dt2) - set(X)

# create zero-filled matrix where the rows are equal to the number
# of row in `test` and columns equal the number of categories missing (i.e. set difference 
# between relevant `train` and `test` columns
feature_difference_df_2 = pd.DataFrame(data=np.zeros((X.shape[0], len(feature_difference_2))),
                                     columns=list(feature_difference_2))

# add "missing" features back to `test
X = X.join(feature_difference_df_2)
X
```

```python
print(X.columns)
print(dt2.columns)
```

```python
#df = df[["C", "A", "B"]]
dt2 = dt2[X.columns]
dt2
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=2021)
```

```python
print("X_train:", X_train.shape) # X_train
print("X_test:",X_test.shape) # X_test
print("y_train:",y_train.shape) # y_train
print("y_test:",y_test.shape) # y_test
```

```python

# gradient boosting for regression in scikit-learn
from numpy import mean
from numpy import std
from sklearn.datasets import make_regression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from matplotlib import pyplot
model = GradientBoostingRegressor()
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(model, X_train, y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1, error_score='raise')
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
# # fit the model on the whole dataset
model = GradientBoostingRegressor()
model.fit(X_train, y_train)
```

```python
n_scores.shape
```

```python
print("Training R2 score is",r2_score(y_train,model.predict(X_train)))
print("Testing R2 score is",r2_score(y_test,model.predict(X_test)))
```
